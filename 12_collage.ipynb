```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
images=os.listdir("./VariousImage/")
img=[]
```

```python
for i in images:
    im=cv2.imread("./VariousImage/"+str(i),0)
    im=cv2.resize(im,(512,512))
    img.append(im)
```

```python
count=1
hor=[]
for i in range(4):
    im1=np.hstack([img[i*4+0],img[i*4+1]])
    im2=np.hstack([img[i*4+2],img[i*4+3]])
    im3=np.hstack([im1,im2])
    hor.append(im3)
im1=np.vstack([hor[0],hor[1]])
im2=np.vstack([hor[2],hor[3]])
final_img=np.vstack([im1,im2])

plt.figure(figsize=(6,6))
plt.xticks([])
plt.yticks([])
plt.imshow(final_img,"gray")
plt.savefig("collage_of_16_images")
```
