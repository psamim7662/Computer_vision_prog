```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Black Dot

```python
black_dot=cv2.imread("CoverImages/output.png",0)
plt.imshow(black_dot,cmap='gray')
plt.xticks([])
plt.yticks([])
```

```python
_,b_black_dot=cv2.threshold(black_dot,100,255,cv2.THRESH_BINARY|cv2.THRESH_OTSU)
contours,_=cv2.findContours(b_black_dot,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
# plt.imshow(b_black_dot,cmap='gray')
dots=[]
for i in contours:
    if 5<cv2.contourArea(i)<100:
        dots.append(i)
print(f"The number of black dots in the image : {len(dots)}")
```
