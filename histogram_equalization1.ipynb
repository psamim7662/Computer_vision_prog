Histogram Equalization

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("CoverImages/Barbara.tif")
img=img[:,:,0]
total=img.size
im=img.copy()
rows=img.shape[0]
print(rows)
```

```python
y=img.ravel()
z=list(y)
hist_temp=[]
# print(hist_temp)
for i in range(0,256):
    hist_temp.append(z.count(i))
    
# total=img.size
```

```python
hist=[]
for i in range(0,256):
    hist.append(hist_temp[i]/total)
cummProb=[]

sum=0
for i in range(0,256):
    sum+=hist[i]
    cummProb.append(round(sum,2))
```

```python
output=[]
for i in range(0,256):
    output.append(round(cummProb[i]*255,0))
```

```python
finalImg=np.zeros((img.shape[0],img.shape[1]),dtype='uint8')
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        finalImg[i][j]=output[img[i,j]]
print(finalImg)
```

```python
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.title("Input Image")
plt.imshow(img,cmap='gray')
plt.subplot(2,2,2)
plt.title("Output Image")
plt.imshow(finalImg,cmap='gray')
plt.show()
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.title("Input Histogram")
plt.hist(img)
plt.subplot(2,2,2)
plt.title("Output Histogram")
plt.hist(finalImg)
plt.show()
```
