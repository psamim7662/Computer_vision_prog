```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Complete Tri-Colour using Matplotlib

```python

a=plt.Rectangle((0,1),height=2,width=12,facecolor='green',edgecolor='grey')
b=plt.Rectangle((0,3),height=2,width=12,facecolor='white',edgecolor='grey')
c=plt.Rectangle((0,5),height=2,width=12,facecolor='#FF9933',edgecolor='grey')
d=plt.Circle((6,4),1,fill=False,color='#000088ff',linewidth=7)

# plt.figure(figsize=(15,25))
j,k=plt.subplots(figsize=(25,15))
k.add_patch(a)
k.add_patch(b)
k.add_patch(c)
k.add_patch(d)


# rows=img.shape[0]
# cols=img.shape[1]
# center=[int(cols/2),int(rows/2)]

radius=1.0
for i in range(0,24):
    p=6+radius/2*np.cos(np.pi*i/12+np.pi/40)
    q=6+radius/2*np.cos(np.pi*i/12-np.pi/40)
    r=4+radius/2*np.sin(np.pi*i/12+np.pi/40)
    s=4+radius/2*np.sin(np.pi*i/12-np.pi/40)
    t=6+radius*np.cos(np.pi*i/12)
    u=4+radius*np.sin(np.pi*i/12)
    k.add_patch(plt.Polygon([[6,4],[p,r],[t,u],[q,s]],fill=True,closed=True,color='#000088ff'))
plt.axis('equal')
plt.savefig('IndianTriColour.jpeg')
plt.show()
```

For making the Tri-colour using opencv-python

```python

img=np.zeros((300,500,3),'uint8')
# img=np.zeros((540,1080,3),'uint8')
```

```python
img[0:100,0:500]=[50,150,255]
img[100:200,0:500]=[255,255,255]
img[200:300,0:500]=[8,136,19]
image=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)




# k.add_patch(a)
```

```python
rows=img.shape[0]
cols=img.shape[1]
center=(int(cols/2),int(rows/2))
xaxis=center[0]
yaxis=center[1]
# i=int(60/8)
y=yaxis
x=xaxis
count=0

cv2.circle(img,center,50,(150,20,10),4)
cv2.circle(img,center,1,(150,20,10),5)


for y in range(yaxis-50-1,yaxis+50+1):
    # x=xaxis-50
    for x in range(xaxis-50-1,xaxis+50+1):  
        # print(x)     
        if((x-xaxis)**2+(y-yaxis)**2==50**2):
            # print(x,y)
            cv2.line(img,center,(x,y),(150,20,10),2)
            count+=1
            
    if(count==24):
        break

print(count)

img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

```
