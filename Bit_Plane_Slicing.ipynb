```python
im=cv.imread("lena_color.tiff",0)
plt.imshow(im,'gray')
plt.xticks([])
plt.yticks([])
plt.title("Gray Image")
plt.show()
```

```python
bit=[]
for i in range(8):
    bitplane=im&(1<<i)
    bit.append(bitplane)
fig=plt.figure(figsize=(10,6))
for i in range(8):
    plt.subplot(2,4,i+1)
    plt.title("Bitplane"+str(i))
    plt.xticks([])
    plt.imshow(bit[i],'gray')
plt.show()
```
