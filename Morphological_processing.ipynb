```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('CoverImages/house.tiff',0)
_,mask=cv2.threshold(img,220,255,cv2.THRESH_BINARY_INV)
cv2.imshow('original',img)
cv2.imshow('Binary',mask)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
img = cv2.imread('CoverImages/peeper.tiff',0)
_,mask=cv2.threshold(img,220,255,cv2.THRESH_BINARY_INV)
kernel=np.ones((3,3),np.uint8)
#dilation
dilation=cv2.dilate(img,mask,iterations=2)
erosion=cv2.erode(img,kernel,iterations=1)
opening=cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)
closing=cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel)
mg=cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
tophat=cv2.morphologyEx(img,cv2.MORPH_TOPHAT,kernel)

plt.figure(figsize=(10,8))
titles=['image','img','dialation','erosion','opening','closing','GRADIENT','TOP HAT']
images=[img,img,dilation,erosion,opening,closing,mg,tophat]
for i in range(8):
    plt.subplot(2,4,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])    
    plt.xticks([]),plt.yticks([])
plt.show()
```
