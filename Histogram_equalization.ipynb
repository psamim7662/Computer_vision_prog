```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img=cv.imread("lena_color.tiff",0)
plt.imshow(img,'gray')
plt.xticks([])
plt.yticks([])
plt.title("Gray Image")
plt.show()
```

```python
# Create a NumPy array of shape (256,) and dtype np.float16 filled with zeros

a=np.zeros((256,),dtype=np.float16)
b=np.zeros((256,),dtype=np.float16)
height,width=img.shape
img.shape
height
width
```

```python
#finding histogram
for i in range(width):
    for j in range(height):
     g=img[j,i]
     a[g]=a[g]+1
print(a) 
```

```python
#performing histogram equalization
tmp = 1.0/(height*width)
b = np.zeros((256,),dtype=np.float16)
tmp
b
```

```python
for i in range(256):
    for j in range(i+1):
        b[i] += a[j] * tmp;
    b[i] = round(b[i] * 255);
```

```python
# b now contains the equalized histogram
b=b.astype(np.uint8)
b
```

```python
#Re-map values from equalized histogram into the image
for i in range(width):
    for j in range(height):
        g = img[j,i]
        img[j,i]= b[g]

cv.imshow('image',img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img= cv.imread('lena_color.tiff', cv.IMREAD_GRAYSCALE)
plt.imshow(img,'gray')
plt.xticks([])
plt.yticks([])
plt.title("Gray Image")
plt.show()
```

```python
# Calculate the histogram
histogram=np.zeros(256,dtype=np.float32)
height,width=img.shape
total_pixels=height*width
height
total_pixels
```

```python
# Compute the histogram
for i in range(height):
    for j in range(width):
        intensity = img[i, j]
        histogram[intensity] += 1
#normalize the histogram
histogram/=total_pixels
```

```python
#cumulative distribution function
cdf=np.zeros(256)
cdf[0]=histogram[0]
for i in range(1,256):
    cdf[i]=cdf[i-1]+histogram[i]
    
```

```python
# Perform histogram equalization
equalized_image = np.zeros_like(img)
for i in range(height):
    for j in range(width):
        intensity = img[i, j]
        equalized_image[i, j] = np.round(cdf[intensity] * 255)
```

```python

plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.title('Original Image')
```

```python
plt.subplot(2, 2, 2)
plt.hist(img.flatten(), 256, [0, 256], color='b')
plt.xticks([])
plt.yticks([])

plt.title('Original Histogram')
```

```python
plt.subplot(2, 2, 3)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')
plt.xticks([])
plt.yticks([])

plt.show()
```
