Global Thresholding

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('CoverImages/lena.tiff', cv2.IMREAD_GRAYSCALE)

```

```python

T = np.mean(image)
G1 = np.where(image > T, 255, 0)  # Pixels with intensity values > T in G1
G2 = np.where(image <= T, 255, 0)  # Pixels with intensity values â‰¤ T in G2

```

```python
# Display the segmented images
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(G1, cmap='gray')
plt.title('Intensity > T')

plt.subplot(1, 3, 3)
plt.imshow(G2, cmap='gray')
plt.title('Intensity â‰¤ T')

plt.show()
```

```python

delta = 0.5

T = np.mean(image)

while True:
    #Segment the image using threshold value 
    fore = image[image > T]  
    back = image[image <= T] 
    m1 = np.mean(fore)
    m2 = np.mean(back)
    new_Threshold = (m1 + m2) / 2

    if abs(new_Threshold - T) < delta:
        break
    T = new_Threshold

#Segment the image as g(x, y) = 1 if f(x, y) > T and g(x, y) = 0 if f(x, y) â‰¤ T
segmented_image = np.where(image > T, 255, 0)



```

```python
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(segmented_image, cmap='gray')
plt.title('output Image')

plt.show()
```
